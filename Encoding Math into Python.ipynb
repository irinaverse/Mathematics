Also, Python can solve equations graphically. We can use many numbers and plot them as a plotting function. Then, we see where our solution lies. These are visual solutions, we see how numbers move or even not move:

```python
def plot_math_functions(function_f, function_g):
    
    x = np.linspace(-3, 5, 1000)

    function_f_vectorized = [np.vectorize(f) for f in function_f]
    y1 = [function_f_vectorized(x) for function_f_vectorized in function_f_vectorized]
    
    function_g_vectorized = [np.vectorize(g) for g in function_g]
    y2 = [function_g_vectorized(x) for function_g_vectorized in function_g_vectorized]

    plt.figure(figsize=(5, 4))
    for y in y1:
        plt.plot(x, y, lw = 1.7, color = "darkgoldenrod")
        plt.xlabel("$x$-axis", fontsize = "8.5")
        plt.ylabel("$y$-axis", fontsize = "8.5")
        plt.legend(["$f(x) = (2x + 3)$", "$g(x) = 0$"], fontsize = "6.5")
        
    for y in y2:
        plt.plot(x, y, lw = 1.8, color = "saddlebrown")
        plt.xlabel("$x$-axis", fontsize = "9")
        plt.ylabel("$y$-axis", fontsize = "9")
        plt.legend(["$f(x) = (2x + 3)$", "$g(x) = 0$"], fontsize = "6.5")

    plt.title("Plotting math functions", fontsize = "9", pad = 5)
    plt.show()
```

```python
plot_math_functions([lambda x: 2*x + 3], [lambda x: 0])
```

The plot shows a function $f(x)$ where it moves through both axes...

And another function $g(x)$ = 0 which moves only through the $x$-axis, $y$ stays the same in the $zero$ plane...

A famous one in mathematics is the Fibonacci sequence. It is a set of numbers in which each number is the sum of the two numbers before it. This sequence has a mathematical pattern that repeats forever and has been studied a lot by mathematicians and computer scientists. 

The Fibonacci sequence is defined as $$F_{(n)} = F_{(n - 1)} + F_{(n - 2)}$$

We encode the formula into Python and have much more opportunities to work with numbers:

```python
def fibonacci_sequence(number):
    
    fibonacci = [0, 1]  # fixed values
    for num in range(2, number + 1):  # so we start from the second index
        fibonacci.append(fibonacci[num - 1] + fibonacci[num - 2])
    return fibonacci
```

```python
fibonacci_sequence(10)   
```

I will share something about vectors from my homework. It is sooo fascinating... two vectors can form a basis in the 2D space and a third vector can be represented as their linear combination. Whaaat?

Let us name:
- the two basis vectors $e_1, e_2$
- the third vector just $v$
- the unknown coefficients $ \lambda_1, \lambda_2 $

And now express them as a linear combination:$$ v = \lambda_1e_1 + \lambda_2e_2 $$

```python
def find_linear_combination_coefficients(e1, e2, v):

    basis_matrix = np.array([e1, e2]).T
    
    return np.linalg.solve(basis_matrix, v)
```

```python
def plot_coeff_vectors(vectors):
    
    start_x = [v[0] for v in vectors]
    start_y = [v[1] for v in vectors]
    
    length_x = [v[2] - v[0] for v in vectors]
    length_y = [v[3] - v[1] for v in vectors] 
    
    # scale is length of vectors (a smaller scale parameter makes the arrow longer)
    plt.quiver(start_x, start_y, length_x, length_y,
               angles = "xy", scale_units = "xy", scale = 1, width = 0.007,
               color = ["tan", "sienna", "gold"])
    
    plt.xlim(-6, 10)
    plt.ylim(-3, 10)
    
    plt.xlabel("x-axis", fontsize = "8")
    plt.ylabel("y-axis", fontsize = "8")
    
    plt.scatter([0, 0], [0, 0], s = 34, color = "darkorange")     # center
    
    plt.show()
```

```python
e1, e2 = [[-3.2, 5.2], [8.2, 6.2]]   # basis vectors

vector = [2.2, 4.2]   # golden vector v (the star)

coefficients = find_linear_combination_coefficients(e1, e2, vector)

print("Coefficients: ", str(coefficients)) 
```

```python
plot_coeff_vectors([[0, 0, i[0], i[1]] for i in [e1, e2, vector]])
```

A tangent line is a straight line that touches a curve at just one point. This important point shows us the behavior of the function and its direction. The right relationship of the elements help us to understand data and numbers. We see the slope of the tangent line as a rate of change - where are things going?

In math language, the rate of change is the derivative of a function $f$ at a point $(x, f(x))$ or $f$ prime of $x$. It is the speed at which a variable changes over a specific period of time. And why the tangent line? The tangent line allows us to find the slope of a curved function.

$$ f'(x) = \lim_{dx \to 0} \frac{f(x + dx) - f(x)}{dx} $$

Let $x = 1$ and $dx$ approach $0$, then...

```python
def f(x):
    return 3 * x**2 - 4 * x                 # f(x) = 3x^2 - 4x
```

```python
def numerical_lim(f, x, dx):
    return (f(x + dx) - f(x)) / dx             # the formula above

x = 1
dx = 0.1
for i in range(5):
    dx *= 0.1
    print(f"dx = {dx:.5f}, numerical limit = {numerical_lim(f, x, dx):.5f}")
```
