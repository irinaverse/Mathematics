In math and science we encounter many situations in which analytical integration of a function or a differential equation of a function is difficult or impossible to solve. In this section I will show how Scientific Python can help through its high level mathematical algorithms. A simple trapezium formula calculates the integral of a function $f(x)$ as the area under the curve representing $f(x)$ by approximating it with the sum of trapeziums and the area of each trapezium is calculated as width times the average height.

$$\int_0^1 \frac{x^4(1 - x)^4}{1 + x^2} dx$$ 

The LaTeX formulas are beautiful, aren't they?! Let's write a small program to evaluate the integral shown above:

```python
def function(x):
    return x**4 * (1 - x)**4 / (1 + x**2)

#define a function to do integration of f(x) between 0 and 1:
def trapezium(function, n):
    h = 1 / float(n)
    integral = 0.5 * h * (function(0) + function(1))
    
    for i in range(1, int(n)):
        integral = integral + h * function(i * h)
    return integral
```

```python
trapezium(function, 100)
```

Question: How many trapeziums are needed to get a relative error of less than 1 part in 1,000,000? Computation follows:

```python
n = 1
while abs(trapezium(function, n) - trapezium(function, n * 2)) > 1e-6:
    n += 1

print("Trapeziums needed: ", int(n))
```

There is art in mathematics indeed. In 1963 Edward Lorenz, an American mathematician and meteorologist, published his famous set of coupled nonlinear first-order ordinary differential equations. They are relative simple but the resulting behavior is wonderfully complex. The equations are:

$$ \frac{dx}{dt} = \sigma(y - x) $$

$$ \frac{dy}{dt} = x(\rho - z) - y $$

$$ \frac{dz}{dt} = xy - \beta z $$

The equations describe the rate of change of three quantities ($dx$, $dy$, $dz$) with respect to time ($dt$):
- $x$ is proportional to the rate of convection
- $y$ is proportional to the horizontal temperature variation
- $z$ is proportional to the vertical temperature variation
- the constants $\sigma$, $\rho$, and $\beta$ are system parameters proportional to the Prandtl number, Rayleigh number and certain physical dimensions of the layer itself.

The solution executes a trajectory, plotted in three dimensions, that winds around and around, neither predictable nor random, occupying a region known as its attractor. Lorenz' work is important because he had an insight of the essence of chaos and found a unimodal structure in the flow of the solution.

```python
def lorenz(xyz, *, s = 10, r = 28, b = 3):
    
    # xyz: array-like, shape(3), point of interest in 3D space.
    # s - sigma (the standard deviation)
    # r - rho (density)
    # b - beta
    
    x, y, z = xyz
    delta_x = s * (y - x)
    delta_y = x * (r - z) - y
    delta_z = x * y - b * z
    return np.array([delta_x, delta_y, delta_z])

dt = 0.01
num_steps = 1000

# Need one more for the initial values
xyzs = np.empty((num_steps + 1, 3))

# Set initial values
xyzs[0] = (0., 1., 1.05)

# Step through "time", calculating the partial derivatives at the current point and using them to estimate the next point
for i in range(num_steps):
    xyzs[i + 1] = xyzs[i] + lorenz(xyzs[i]) * dt
    
ax = plt.figure().add_subplot(projection = "3d")

ax.plot(*xyzs.T, lw = 1.1, color = "black", alpha = 0.8)
ax.set_xlabel("x-axis", fontsize = "9")
ax.set_ylabel("y-axis", fontsize = "9")
ax.set_zlabel("z-axis", fontsize = "9")
ax.set_title("Lorenz Attractor", fontsize = "10", loc = "right")

plt.show()
```

Edward Lorenz discovered what chaos really is - small perturbations in the atmosphere can cause enormous climate changes, and mentioned for the first time his famous poetic sentence "a butterfly flapping its wings in Brazil can produce a tornado in Texas" in December 1972 at the AAAS.
